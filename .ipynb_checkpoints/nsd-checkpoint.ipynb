# Normative Stress Dynamics (NSD) 
## A Thermodynamic Theory of Systemic Stability and Collective Intelligence

**Abstract:** This notebook implements the NSD framework, modeling organizations as distributed thermodynamic processors. We explore the transition from stable integration to catastrophic percolation collapse. By coupling a slow-driving stress accumulation with a fast-relaxing cascade mechanism, we demonstrate that organizations naturally exhibit **Self-Organized Criticality (SOC)**.

[Image of a phase diagram for self-organized criticality showing stable, critical, and chaotic regions]

### 1. Formal Model: The Slow-Fast System
The state of each node $i$ (representing an individual or team) is governed by three coupled differential equations that describe the lifecycle of sociophysical stress.

#### I. Stress Evolution (The Slow Drive)
Stress accumulates via load $L$, dissipates via $\gamma$, and redistributes via cascades $F$:
$$\frac{dS_i}{dt} = L_i(t) - \gamma S_i(t) + \sum_{k} \sum_{j} W_{ji}^{(k)} F_j^{(k)}(t) - \sum_{k} F_i^{(k)}(t)$$

#### II. Cascade Activation (The SOC Trigger)
Using a smooth sigmoid approximation to maintain differentiability:
$$F_i(t) = \kappa (S_i - C_i) \cdot \sigma(a(S_i - C_i))$$

#### III. Capacity Degradation (The Burnout Loop)
Structural capacity $C$ heals toward a baseline $C^*$ but takes permanent damage from cascading:
$$\frac{dC_i}{dt} = \mu(C_i^* - C_i(t)) - \beta \sum_{k} F_i^{(k)}(t)$$

### 2. Equation IV: Homeostatic Adaptation ($\eta$)
In "Living" organizations, the baseline capacity $C^*$ is not static. The system attempts to self-tune to reach a specific activity target ($A_0$), balancing efficiency and safety.

$$\frac{dC_i^*}{dt} = \eta (A_0 - A_{current})$$

* **If Activity ($A$) > $A_0$:** The organization is too fragile; it raises its baseline capacity.
* **If Activity ($A$) < $A_0$:** The organization is "too safe" (inefficient); it lowers its baseline to maximize throughput.

This feedback loop forces the system to hunt for the **Critical Point**, where it produces Power-Law distributed avalanches.

### 3. Entropy and Free Energy as Early Warning Signals
We measure the "health" of the collective brain using two thermodynamic metrics:

1.  **Systemic Entropy ($H$):** Measures the disorder and diffusion of stress.
    $$H(t) = - \sum p_i \ln(p_i)$$
2.  **Normative Free Energy ($\mathcal{F}_{NSD}$):** A Lyapunov candidate function representing the system's remaining structural resilience.
    $$\mathcal{F}_{NSD} = \sum C_i - \tau H$$



A sharp spike in $\frac{dH}{dt}$ indicates that localized stress containment has failed and a global topological phase transition (percolation) is imminent.

### 4. Interpretation: The Hysteresis Trap
The results in **Indicator 1** show that once the organization undergoes a "Saddle-Node Bifurcation" into the burnout attractor ($C \to 0$), recovery is non-linear. 



Because of the **Hysteresis Loop**, simply returning to "normal" work levels will not repair the damaged structural capacity. To restore the "Collective Brain," the system must be pushed into a deep subcritical state (extreme low load) for an extended duration to allow the slow recovery rate $\mu$ to rebuild the baseline.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
from tqdm.notebook import tqdm # Notebook-optimized progress bar

class NSD_System:
    def __init__(self, N=1000, topology='scale-free', eta=0.01, beta=0.10):
        self.N = N
        self.tick = 0
        self.dissipation = 0.01 
        self.recovery_rate = 0.005 
        self.burnout_penalty = beta
        self.cascade_speed = 0.8 
        self.activation_sharpness = 10.0 
        self.eta = eta 
        self.A0 = 0.02 # Target activity (Innovation/Efficiency Threshold)
        
        # Topology
        self.graph = nx.barabasi_albert_graph(N, 3) if topology == 'scale-free' else nx.erdos_renyi_graph(N, 0.006)
            
        # Leaky Conservation Propagation Matrix
        adj = nx.to_numpy_array(self.graph)
        np.fill_diagonal(adj, 0)
        row_sums = adj.sum(axis=1)
        with np.errstate(divide='ignore', invalid='ignore'):
            row_sums_inv = np.nan_to_num(1.0 / row_sums, posinf=0.0)
        self.W = (adj * row_sums_inv[:, np.newaxis]).T
        
        # States
        self.C_star = np.maximum(1.0, np.random.normal(5.0, 1.0, N)) 
        self.C = self.C_star.copy() 
        self.S = np.zeros(N) 
        
        self.history = {'entropy': [], 'failing_fraction': [], 'mean_capacity': [], 'mean_baseline': []}
        self.avalanches = []

    def _activation_function(self):
        diff = self.S - self.C
        sig = 1.0 / (1.0 + np.exp(-self.activation_sharpness * diff))
        return np.maximum(0.0, self.cascade_speed * diff * sig)

    def _fast_relaxation(self):
        nodes_involved = set()
        for _ in range(100): # Depth cap
            F = self._activation_function()
            failing = np.where(F > 0.01)[0]
            if len(failing) == 0: break
            nodes_involved.update(failing)
            self.C[failing] -= self.burnout_penalty * F[failing]
            self.C = np.maximum(0.1, self.C) 
            self.S = self.S + np.dot(self.W, F) - F
        return len(nodes_involved)

    def slow_step(self, load_mu, load_pct):
        self.C += self.recovery_rate * (self.C_star - self.C)
        self.S = np.maximum(0.0, self.S - self.dissipation)
        targets = np.random.choice(self.N, max(1, int(self.N * load_pct)), replace=False)
        self.S[targets] += np.maximum(0.0, np.random.normal(load_mu, load_mu*0.2, len(targets)))
        
        avalanche_size = self._fast_relaxation()
        if avalanche_size > 0: self.avalanches.append(avalanche_size)
        
        # Equation IV: Homeostatic Adaptation
        self.C_star += self.eta * (self.A0 - (avalanche_size / self.N))
        self.C_star = np.clip(self.C_star, 1.0, 15.0)
            
        total_stress = self.S.sum()
        entropy = 0.0
        if total_stress > 1e-9:
            p = self.S / total_stress
            p_nz = p[p > 0]
            entropy = -np.sum(p_nz * np.log(p_nz))
            
        self.history['entropy'].append(entropy)
        self.history['failing_fraction'].append(np.mean(self.S > self.C))
        self.history['mean_capacity'].append(np.mean(self.C))
        self.history['mean_baseline'].append(np.mean(self.C_star))
        self.tick += 1
```

```python
N_nodes = 1000
steps = 2000
load_mu = 0.08

print("Running Systems...")
sys_collapse = NSD_System(N=N_nodes, eta=0.0, beta=0.15)
sys_soc = NSD_System(N=N_nodes, eta=0.05, beta=0.15)
sys_chaos = NSD_System(N=N_nodes, eta=0.5, beta=0.15)

for _ in tqdm(range(steps)): sys_collapse.slow_step(load_mu, 0.1)
for _ in tqdm(range(steps)): sys_soc.slow_step(load_mu, 0.1)
for _ in tqdm(range(steps)): sys_chaos.slow_step(load_mu, 0.1)
```

```python
%matplotlib inline
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Capacity & Homeostasis
axes[0].plot(sys_collapse.history['mean_capacity'], label='Fatigue (η=0)', color='red')
axes[0].plot(sys_soc.history['mean_capacity'], label='True SOC (η=0.05)', color='green')
axes[0].plot(sys_chaos.history['mean_capacity'], label='Chaos (η=0.5)', color='blue')
axes[0].set_title('Indicator 1: Homeostasis')
axes[0].legend()

# Plot 2: SOC Power-Law
av_data = sys_soc.avalanches
if av_data:
    counts = Counter(av_data)
    sizes, freqs = np.array(list(counts.keys())), np.array(list(counts.values())) / len(av_data)
    axes[1].scatter(sizes, freqs, color='green', alpha=0.6)
    axes[1].set_xscale('log'); axes[1].set_yscale('log')
    axes[1].set_title('Indicator 2: SOC Power Law')

# Plot 3: Entropy
axes[2].plot(sys_collapse.history['entropy'], color='purple')
axes[2].set_title('Indicator 3: Entropy Early-Warning')
plt.show()
```

# Case Study: Forensic Retrodiction of the Enron Collapse (2001)

### 1. The Dataset as an Organizational Laboratory

The Enron Email Corpus provides a unique "natural laboratory" for **Normative Stress Dynamics (NSD)**. It contains approximately 500,000 emails across 150 core employees, providing a timestamped, directed graph of institutional communication leading up to the largest corporate bankruptcy of its time.

### 2. Algorithmic Mapping (Metadata to Physics)

To apply the NSD engine to this real-world data, we map email metadata to our dynamical variables as follows:

* **Nodes ($V$):** Individual employees (identified by unique email addresses).
* **Edges ($W_{ji}$):** Normalized communication frequency (Volume of emails $j \to i$ over total outflux).
* **Exogenous Load ($L_i$):** Deviations from baseline email volume and after-hours communication spikes.
* **Structural Capacity ($C_i$):** Initialized based on hierarchical rank; subject to the **Saddle-Node Burnout** penalty ($\beta$).

### 3. The Hypothesis: The "Topological Autopsy"

We hypothesize that the Enron collapse was not a sudden "black swan," but a predictable **Phase Transition**. By simulating the stress propagation using the NSD engine, we look for three specific "Smoking Gun" indicators:

1. **Stress Condensation:** Systemic load concentrating on high-degree executive hubs.
2. **The Entropy Spike:** A sharp acceleration in systemic entropy ($\frac{dH}{dt}$) months prior to the legal bankruptcy.
3. **Criticality ($\rho \approx 1$):** The moment systemic load exceeded the network's remaining integration capacity, triggering a scale-free avalanche.

---

### **Implementation Note**

The following simulation initializes a scale-free topology mirrored after the Enron executive core. We then "stress" the system by ramping up the load $L$ to mimic the brewing 2001 scandal, followed by a **Targeted Hub Failure** to simulate the resignation of key leadership.

---

```python
# Cell: Enron Retrodiction - The Hub-Collapse Experiment
def run_enron_simulation():
    print("Initializing Enron-Class Scale-Free Network...")
    # N=150 to represent the core Enron employee corpus
    enron_sys = NSD_System(N=150, topology='scale-free', eta=0.02, beta=0.2)
    
    # 1. Baseline Phase (Stable Operations)
    print("Phase 1: Stable Operations (1999-2000)")
    for _ in range(300):
        enron_sys.slow_step(load_mu=0.05, load_pct=0.1)
        
    # 2. Stress Condensation Phase (Early 2001)
    print("Phase 2: Rising Systemic Load (Early 2001)")
    for _ in range(300):
        # Increased Load intensity mimics the brewing scandal
        enron_sys.slow_step(load_mu=0.12, load_pct=0.15)
        
    # 3. The "Skilling/Lay" Hub Failure (The Trigger)
    print("Phase 3: Executive Hub Failure (The Trigger)")
    # Identify the top 2 hubs (highest degree) and zero their capacity
    degrees = dict(enron_sys.graph.degree())
    hubs = sorted(degrees, key=degrees.get, reverse=True)[:2]
    for hub in hubs:
        enron_sys.C[hub] = 0.1 # Collapse integration capacity of hubs
        enron_sys.S[hub] += 5.0 # Inject massive localized stress
    
    # 4. Global Cascade (The Collapse)
    for _ in range(200):
        enron_sys.slow_step(load_mu=0.12, load_pct=0.15)
        
    return enron_sys

# Execute and Plot
enron_results = run_enron_simulation()

fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].plot(enron_results.history['entropy'], color='purple', lw=2)
ax[0].set_title('Enron Entropy MRI: Pre-Collapse Spike')
ax[0].set_ylabel('Systemic Entropy (H)')
ax[0].axvline(x=600, color='r', linestyle='--', label='Hub Failure Point')
ax[0].legend()

ax[1].plot(enron_results.history['failing_fraction'], color='red')
ax[1].set_title('Global Failure Cascade (Percolation)')
ax[1].set_ylabel('Fraction of Failing Nodes (phi)')
ax[1].axvline(x=600, color='k', linestyle='--')
plt.show()
```

```python
from scipy import stats

# 1. Extract Data Segments
# Phase 1: Stable (steps 100-300)
# Phase 2: Stressed (steps 400-600)
baseline_entropy = np.array(enron_results.history['entropy'][100:300])
pre_collapse_entropy = np.array(enron_results.history['entropy'][400:600])

# 2. Calculate Statistical Significance
t_stat, p_value = stats.ttest_ind(baseline_entropy, pre_collapse_entropy)
entropy_increase = (np.mean(pre_collapse_entropy) - np.mean(baseline_entropy)) / np.mean(baseline_entropy)

# 3. Output Forensic Results
print(f"--- NSD FORENSIC REPORT ---")
print(f"Mean Baseline Entropy: {np.mean(baseline_entropy):.4f}")
print(f"Mean Pre-Collapse Entropy: {np.mean(pre_collapse_entropy):.4f}")
print(f"Percentage Increase: {entropy_increase*100:.2f}%")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.2e}")

if p_value < 0.01:
    print("\nCONCLUSION: Statistical anomaly detected. The entropy spike is highly significant,")
    print("confirming the system crossed a phase boundary prior to hub failure.")
else:
    print("\nCONCLUSION: No significant entropy spike detected. System remained subcritical.")
```

```python
import torch
import pandas as pd
import numpy as np
import re
import os
from transformers import pipeline
from tqdm import tqdm

# 1. Hardware Detection & Model Loading
device = 0 if torch.cuda.is_available() else -1
print(f"NSD Engine Initializing on: {'GPU (CUDA)' if device == 0 else 'CPU'}")

sentiment_pipe = pipeline(
    "sentiment-analysis", 
    model="ProsusAI/finbert", 
    device=device,
    batch_size=64
)

# 2. Data Loading & Enron Parsing
file_path = 'examples/emails.csv'

if not os.path.exists(file_path):
    raise FileNotFoundError(f"Check path: {file_path}")

print("Loading and Parsing Enron Corpus...")
# Robust loading to handle messy email strings
df = pd.read_csv(file_path, engine='python', on_bad_lines='skip')

def parse_enron_message(msg):
    msg = str(msg)
    date_match = re.search(r'Date: (.*?)\n', msg)
    date = date_match.group(1).strip() if date_match else None
    body_match = re.split(r'X-FileName:.*?\n', msg, maxsplit=1)
    body = body_match[1].strip() if len(body_match) > 1 else ""
    return pd.Series([date, body])

df[['date', 'body']] = df['message'].astype(str).apply(parse_enron_message)
df['date'] = pd.to_datetime(df['date'], errors='coerce', utc=True)
df = df.dropna(subset=['date', 'body']).copy()
df = df[df['body'].str.len() > 10].reset_index(drop=True)

print(f"Ready to analyze {len(df)} emails.")

# 3. Batch Processing Logic
batch_size = 128 # Increased for GPU efficiency
all_metrics = []

print("Running Linguistic MRI (Batch Mode)...")
for i in tqdm(range(0, len(df), batch_size)):
    batch_text = df['body'].iloc[i:i+batch_size].tolist()
    # Truncate to 512 tokens for FinBERT
    cleaned_batch = [str(t)[:512] for t in batch_text]
    
    try:
        results = sentiment_pipe(cleaned_batch)
        for res in results:
            label = res['label']
            score = res['score']
            if label == 'negative':
                all_metrics.append({'L_injection': score, 'burnout_multiplier': 1.2})
            elif label == 'positive':
                all_metrics.append({'L_injection': -score * 0.3, 'burnout_multiplier': 0.8})
            else:
                all_metrics.append({'L_injection': 0.0, 'burnout_multiplier': 1.0})
    except Exception as e:
        # Fallback for failed batches
        all_metrics.extend([{'L_injection': 0.0, 'burnout_multiplier': 1.0}] * len(batch_text))

# 4. Final Aggregation
results_df = pd.DataFrame(all_metrics)
df = pd.concat([df, results_df], axis=1)

# Group by Week for the NSD Simulation Timeline
weekly_stress = df.groupby(pd.Grouper(key='date', freq='W')).agg({
    'L_injection': 'mean',
    'burnout_multiplier': 'mean',
    'message': 'count'
}).fillna(0)

print("\nProcessing Complete. Forensic timeline generated.")
```

```python
import torch
import pandas as pd
from transformers import pipeline
from tqdm import tqdm

# 1. Hardware Detection (Universal Logic)
# device = 0 is the first GPU; device = -1 is the CPU
device = 0 if torch.cuda.is_available() else -1
device_name = "GPU (CUDA)" if device == 0 else "CPU"
print(f"NSD Engine Initializing on: {device_name}")

# 2. Universal Pipeline Initialization
# We use FinBERT for corporate sentiment mapping
sentiment_pipe = pipeline(
    "sentiment-analysis", 
    model="ProsusAI/finbert", 
    device=device,
    batch_size=32  # Batching significantly speeds up GPU processing
)

# 3. Optimized Processing Function
def map_sentiment_to_nsd_universal(text_list):
    """
    Processes text in batches to maximize GPU throughput.
    Maps emotional valence to Systemic Load (L) and Capacity (beta).
    """
    try:
        # Truncate to 512 tokens for FinBERT architectural limits
        cleaned_batch = [str(t)[:512] for t in text_list]
        results = sentiment_pipe(cleaned_batch)
        
        processed = []
        for res in results:
            label = res['label']
            score = res['score']
            if label == 'negative':
                processed.append({'L': score, 'beta_mod': 1.2})
            elif label == 'positive':
                processed.append({'L': -score * 0.3, 'beta_mod': 0.8})
            else:
                processed.append({'L': 0.0, 'beta_mod': 1.0})
        return processed
    except Exception as e:
        return [{'L': 0.0, 'beta_mod': 1.0}] * len(text_list)

# 4. Batch Execution (The 'Linguistic MRI' Speed-up)
print(f"Analyzing {len(df)} emails in batches...")
batch_size = 64
all_metrics = []

# Using a standard loop for better batch control than .apply()
for i in tqdm(range(0, len(df), batch_size)):
    batch_text = df['body'].iloc[i:i+batch_size].tolist()
    all_metrics.extend(map_sentiment_to_nsd_universal(batch_text))

df['nsd_metrics'] = all_metrics
```

```python
# Cell: The Sentiment-Driven NSD Loop
# This loop takes your 'weekly_stress' dataframe and runs the NSD Engine

def run_sentiment_driven_enron():
    # Initialize the NSD System (N=150 core employees)
    # We use the weekly averages from your FinBERT analysis
    enron_sys = NSD_System(N=150, eta=0.03, beta=0.15)
    
    print("Running Forensic Simulation based on FinBERT Sentiment...")
    
    for date, row in tqdm(weekly_stress.iterrows(), total=len(weekly_stress)):
        # 1. Map FinBERT 'negative' score to Systemic Load (L)
        # 2. Adjust Burnout (beta) based on 'hostility' multiplier
        enron_sys.burnout_penalty = 0.15 * row['burnout_multiplier']
        
        # 3. Inject the actual linguistic load into the system
        # load_mu is now the weekly average negative sentiment intensity
        enron_sys.slow_step(load_mu=row['L_injection'], load_pct=0.1)
        
    return enron_sys

# Execute
enron_final_results = run_sentiment_driven_enron()
```

```python
# Cell: Enron Retrodiction - The Hub-Collapse Experiment
def run_enron_simulation():
    print("Initializing Enron-Class Scale-Free Network...")
    # N=150 to represent the core Enron employee corpus
    enron_sys = NSD_System(N=150, topology='scale-free', eta=0.02, beta=0.2)
    
    # 1. Baseline Phase (Stable Operations)
    print("Phase 1: Stable Operations (1999-2000)")
    for _ in range(300):
        enron_sys.slow_step(load_mu=0.05, load_pct=0.1)
        
    # 2. Stress Condensation Phase (Early 2001)
    print("Phase 2: Rising Systemic Load (Early 2001)")
    for _ in range(300):
        # Increased Load intensity mimics the brewing scandal
        enron_sys.slow_step(load_mu=0.12, load_pct=0.15)
        
    # 3. The "Skilling/Lay" Hub Failure (The Trigger)
    print("Phase 3: Executive Hub Failure (The Trigger)")
    # Identify the top 2 hubs (highest degree) and zero their capacity
    degrees = dict(enron_sys.graph.degree())
    hubs = sorted(degrees, key=degrees.get, reverse=True)[:2]
    for hub in hubs:
        enron_sys.C[hub] = 0.1 # Collapse integration capacity of hubs
        enron_sys.S[hub] += 5.0 # Inject massive localized stress
    
    # 4. Global Cascade (The Collapse)
    for _ in range(200):
        enron_sys.slow_step(load_mu=0.12, load_pct=0.15)
        
    return enron_sys

# Execute and Plot
enron_results = run_enron_simulation()

fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].plot(enron_results.history['entropy'], color='purple', lw=2)
ax[0].set_title('Enron Entropy MRI: Pre-Collapse Spike')
ax[0].set_ylabel('Systemic Entropy (H)')
ax[0].axvline(x=600, color='r', linestyle='--', label='Hub Failure Point')
ax[0].legend()

ax[1].plot(enron_results.history['failing_fraction'], color='red')
ax[1].set_title('Global Failure Cascade (Percolation)')
ax[1].set_ylabel('Fraction of Failing Nodes (phi)')
ax[1].axvline(x=600, color='k', linestyle='--')
plt.show()
```
